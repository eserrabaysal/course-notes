```python
import numpy as np
from sklearn.preprocessing import StandardScaler

degerler = np.array([3, 78, 22, 50, 45, 64])
degerler = degerler.reshape(-1, 1)

ss = np.std(degerler)
ortalama = np.mean(degerler)
print((degerler[0][0] - ortalama) / ss)

sc = StandardScaler()
print(sc.fit_transform(degerler))
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler

degerler = np.array([3, 78, 22, 50, 45, 64])
degerler = degerler.reshape(-1,1)

ss = np.std(degerler)
ortalama = np.mean(degerler)

for i in range(6):
    print((degerler[i][0] - ortalama)/ss)

sc = StandardScaler()
print(sc.fit_transform(degerler))
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

degerler = np.array([3, 78, 22, 50, 45, 64])
xmin = np.min(degerler)
xmax = np.max(degerler)

for i in range(6):
    print((degerler[i] - xmin) / (xmax - xmin))

print(MinMaxScaler().fit_transform(degerler.reshape(-1, 1)))
```

```python
from sklearn.preprocessing import RobustScaler
import numpy as np

degerler = np.array([3, 78, 22, 50, 45, 64])
x75 = np.percentile(degerler, 75)
x25 = np.percentile(degerler, 25)
iqr = x75 - x25

print(x75, x25, iqr, np.median(degerler))

olcDeg = (degerler[0] - np.median(degerler)) / iqr
print(olcDeg)

print(RobustScaler().fit_transform(degerler.reshape(-1, 1)))
```

```python
import numpy as np
from sklearn.preprocessing import Normalizer

# Örnek veri kümesi (2 boyutlu)
veri = np.array([[4, 3, 2, 1],
                 [10, 8, 6, 4],
                 [1, 2, 3, 4]])

# 4. Adım: sklearn kullanarak Normalizer işlemi
normalizer = Normalizer(norm='l2')  # L2 normu kullanılıyor
normalizasyon_sklearn = normalizer.fit_transform(veri)

print("\nsklearn Normalizer Sonucu:\n", normalizasyon_sklearn)
```

```python
import numpy as np
from sklearn.preprocessing import OneHotEncoder,LabelEncoder,OrdinalEncoder

# Örnek kategorik veri
veri = np.array([['Kırmızı', 'Küçük', 'İlkokul'],
                ['Mavi', 'Orta', 'Lise'],
                ['Yeşil', 'Büyük', 'Üniversite'],
                ['Kırmızı', 'Orta', 'Lise']])

# OneHotEncoder örneği (renk için)
onehot_encoder = OneHotEncoder()
renk_encoded = onehot_encoder.fit_transform(veri[:, 0].reshape(-1, 1)).toarray()

# LabelEncoder örneği (boyut için)
label_encoder = LabelEncoder()
boyut_encoded = label_encoder.fit_transform(veri[:, 1])

# OrdinalEncoder örneği (eğitim seviyesi için)
ordinal_encoder = OrdinalEncoder(categories=[['İlkokul', 'Lise', 'Üniversite']])
egitim_encoded = ordinal_encoder.fit_transform(veri[:, 2].reshape(-1, 1))

print(renk_encoded)
print(boyut_encoded)
print(egitim_encoded)
```
