```python
import numpy as np

# Verileri tanımlama (12 ay x 3 bölge)
data = np.array([
    [10, 8, 12],  # Ocak
    [12, 9, 14],  # Şubat
    [15, 11, 16], # Mart
    [18, 13, 18], # Nisan
    [20, 15, 20], # Mayıs
    [22, 17, 22], # Haziran
    [25, 19, 24], # Temmuz
    [28, 18, 28], # Ağustos
    [31, 16, 31], # Eylül
    [19, 14, 19], # Ekim
    [17, 12, 17], # Kasım
    [14, 10, 15]  # Aralık
])

# Ay isimleri
aylar = ["Ocak", "Şubat", "Mart", "Nisan", "Mayıs", "Haziran",
         "Temmuz", "Ağustos", "Eylül", "Ekim", "Kasım", "Aralık"]

# 1. Dizinin boyutunu ve eleman sayısını inceleme
print("1. Dizinin boyutu:", data.shape)
print("   Eleman sayısı:", data.size)

# 2. Her bölgenin yıllık toplam satışı
bolge_toplamlar = np.sum(data, axis=0)
print("\n2. Her bölgenin yıllık toplam satışı:")
for i in range(3):
    print(f"   Bölge {i+1}: {bolge_toplamlar[i]} milyon TL")

# 3. Tüm bölgelerin yıllık ortalama satışı
genel_ortalama = np.mean(bolge_toplamlar)
print(f"\n3. Tüm bölgelerin yıllık ortalama satışı: {genel_ortalama:.2f} milyon TL")

# 4. En yüksek ve en düşük aylık satışlar
max_satis = np.max(data)
min_satis = np.min(data)
max_indeks = np.where(data == max_satis)
min_indeks = np.where(data == min_satis)
print("\n4. En yüksek ve en düşük aylık satışlar:")
for i in range(len(max_indeks[0])):
    print(f"   En yüksek: Bölge {max_indeks[1][i]+1}, {aylar[max_indeks[0][i]]} ayında {max_satis} milyon TL")
for i in range(len(min_indeks[0])):
    print(f"   En düşük: Bölge {min_indeks[1][i]+1}, {aylar[min_indeks[0][i]]} ayında {min_satis} milyon TL")

# 5. En çok ve en az satış yapılan aylar
aylik_toplamlar = np.sum(data, axis=1)
max_ay_indeks = np.argmax(aylik_toplamlar)
min_ay_indeks = np.argmin(aylik_toplamlar)
print(f"\n5. En çok satış yapılan ay: {aylar[max_ay_indeks]} ({aylik_toplamlar[max_ay_indeks]} milyon TL)")
print(f"   En az satış yapılan ay: {aylar[min_ay_indeks]} ({aylik_toplamlar[min_ay_indeks]} milyon TL)")

# 6. Her bölgenin satışlarındaki standart sapma
std_sapma = np.std(data, axis=0)
print("\n6. Her bölgenin standart sapması:")
for i in range(3):
    print(f"   Bölge {i+1}: {std_sapma[i]:.2f} milyon TL")

# 7. Verileri yeniden şekillendirme
print("\n7. Satış verilerinin farklı formatı:")
print(data.reshape(-1))  # Tek boyutlu düzleştirilmiş dizi

# 8. 15 milyon TL'nin üzerindeki satışlar
print("\n8. 15 milyon TL'nin üzerindeki satışlar:")
for i in range(len(aylar)):
    for j in range(3):
        if data[i, j] > 15:
            print(f"   {aylar[i]}, Bölge {j+1}: {data[i, j]} milyon TL")

# 9. Her bölgenin aylık kümülatif satış toplamları
print("\n9. Her bölgenin aylık kümülatif satış toplamları:")
for j in range(3):
    kumulatif = np.cumsum(data[:, j])
    print(f"   Bölge {j+1}: {kumulatif}")

# 10. Her bölgenin medyanı
medyanlar = np.median(data, axis=0)
print("\n10. Her bölgenin satışlarının medyanı:")
for i in range(3):
    print(f"    Bölge {i+1}: {medyanlar[i]} milyon TL")

# 11. Her bölgenin varyansı
varyanslar = np.var(data, axis=0)
print("\n11. Her bölgenin satışlarının varyansı:")
for i in range(3):
    print(f"    Bölge {i+1}: {varyanslar[i]:.2f} (milyon TL)²")

# 12. 25. ve 75. yüzdelik dilimler
yuzde_25 = np.percentile(data, 25)
yuzde_75 = np.percentile(data, 75)
print(f"\n12. Tüm verilerin 25. yüzdelik dilimi: {yuzde_25} milyon TL")
print(f"    Tüm verilerin 75. yüzdelik dilimi: {yuzde_75} milyon TL")

# 13. Her bölgenin en çok satış yaptığı ay
print("\n13. Her bölgenin en çok satış yaptığı ay:")
for j in range(3):
    max_ay = np.argmax(data[:, j])
    print(f"    Bölge {j+1}: {aylar[max_ay]} ({data[max_ay, j]} milyon TL)")

# 14. Satış verilerinden rastgele 5 eleman seçimi
rastgele_elemanlar = np.random.choice(data.flatten(), 5, replace=False)
print("\n14. Satış verilerinden rastgele 5 eleman:")
print("    ", rastgele_elemanlar)

# 15. Satış verilerini düzleştirme
duzlestirme = data.flatten()
print("\n15. Düzleştirilmiş veri:")
print("    ", duzlestirme)
```
